## Part 1

```python
import numpy as np
import matplotlib.pyplot as plt

z = 0

x = np.linspace(-2,2, 1000)
y = np.linspace(-2,2, 1000)



re, im = np.meshgrid(x,y)


c = re + 1j*im

for k in x:
    z = z**2 + c

boundary = np.abs(z) < 2

plt.figure(figsize=(10,8))

plt.imshow(boundary, extent=[-2,2,-2,2], origin='lower', cmap='viridis')

plt.xlabel("Real")
plt.ylabel("Imaginary")
plt.title("Mandelbrot Set")
plt.colorbar()
plt.xlim(-2,0.5)
plt.ylim(-1.25,1.25)

plt.show()
```
