# Part 1

```python
def f(x):
    """
    This function takes any number in the form of an int or a float
    and returns the evaluated polynomial of that number 
    where the polynomial is x**3 - x**2 - 1
    
    Parameters:
    ------------
    x - float or int
    
    Returns:
    f(x) - as a float
    
    """
    return x**3 - x**2 -1


def df(x):
    
    """
    This function takes any number in the form of an int or a float
    and returns the derivative of the polynomial f(x) at that number
    
    Parameters:
    ------------
    x - float or int
    
    Returns:
    f(x) as a float or int
    
    """
    return 2*x**3 - 2*x
```
